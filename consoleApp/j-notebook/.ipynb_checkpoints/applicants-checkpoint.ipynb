```python
file_name = "bsvr_noktalari.xlsx"
data = pd.read_excel (r'../caseData/bsvr_noktalari.xlsx') 
df = pd.DataFrame(data)

df = df.rename(columns={'Kişi Adı': 'applicantName', 'Adresi': 'ad'})
```

```python
df
```

```python


for index, row in df.iterrows():
    g = geocoder.arcgis(row['Adresi'] + row['Şehri'])
    keys = list(g.json.values())
    df.at[index, 'lat'] = keys[3]
    df.at[index, 'lng'] = keys[4]
```

```python
df
```

```python
applicantGdf= gpd.GeoDataFrame(df, geometry= gpd.points_from_xy(df.lng, df.lat))
applicantGdf=applicantGdf.drop(columns=['lng','lat'])

applicantGdf
```

```python
applicantGdf.geometry.crs = {'init':'epsg:4326'}
```

```python
applicantGdf.to_postgis(name="applicants", con=engine, schema="public")
```
